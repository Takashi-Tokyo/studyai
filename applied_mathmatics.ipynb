## 100文字以上で要点のまとめ  

・線形代数とは簡単に言えば、「ベクトルと線形写像の理論」である。また線形写像とは、ベクトルの1次結合を保つ写像のことである。  
　核(kernel),像(image), 階数(rank)は線形写像の一側面としての概念である。  
・行列とはスカラーを表にしたものであり、ベクトルを並べたものである。ベクトルの変換に用いる。  
・線形写像の核は、連立方程式の解ととらえることができ、核を表すことが、連立方程式を解くことになる。  
・線形写像の像は、連立方程式の解が存在する条件を決めることになる。したがって、行列の列ベクトルの中から、その像の基底を求める。  
・線形写像の階数は、線形写像の次元が線形写像の像の次元であり、この次元は、線形写像を行列で表現したとき、その行列の階数に一致する。  
・任意の点Pをそれ自身に移す変換を恒等変換という。    
  点$P(x,y)$から点$P(x',y')$への恒等変換は以下の式で表すことができる。　　
$$\begin{pmatrix}
  x'\\
  y' \\
\end{pmatrix}=
   \begin{pmatrix}
  1&0\\
  0&1 \\
\end{pmatrix}
  \begin{pmatrix}
  x'\\
  y' \\
\end{pmatrix}$$
・上の行列部分$ \begin{pmatrix}
  1&0\\
  0&1 \\
\end{pmatrix}$
 を単位行列という。単位行列は$E$で表す。  
・任意の正則行列$A$について、$A$から単位行列$E$を作る行基本変形を、単位行列$E$に施すと、$A$の逆行列$A^{-1}$が得られる。  
・行列式が0である場合は逆行列を持たない。  
・固有値・固有ベクトルの定義  
・ゼロでないベクトル$v \in \mathbb{C}^n$に対して$Av=\lambda v$となるような複素数 $\lambda \in \mathbb{C}$が存在するとき、$\lambda$を$A$の固有値、$v$を固有値が$\lambda$の固有ベクトルという。  
・固有値・固有ベクトルの求め方
$$
A\vec{x} = \lambda \vec{x} \\
(A - \lambda I) \vec{x} = \vec{0} \\
\vec{x} \neq \vec{0} 　より \\
$$

$$
|A - \lambda I| = 0 \\
\begin{vmatrix}
1 - \lambda & 4 \\
2 & 3 - \lambda
\end{vmatrix} = 0 \\
(1 - \lambda)(3 - \lambda) - 4·2 = 0 \\
\lambda = 5~or -1 \\
$$

$$
\begin{pmatrix}
1 & 4 \\
2 & 3
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
= 5 \begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
よって x_1 = x_2 \\
\begin{pmatrix}
1 & 4 \\
2 & 3
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
= -1
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
よって x_1 = -2x_2
$$
したがって  
$$
\lambda = 5 のとき \vec{x} = \begin{pmatrix}
1 \\
1 \end{pmatrix} の定数倍  \\
\lambda = -1 のとき \vec{x} = \begin{pmatrix}
2 \\
-1 \end{pmatrix} の定数倍
$$
・固有値分解は、
$$
A=V \Lambda V^{-1}
$$
任意の正方行列を上述のような3つの行列の積に変換することを固有値分解という。
$\Lambda$は、行列Aの固有値を対角線上に並べた行列で、$V$は、それに対応する固有ベクトルを並べた行列である。この変換によって、行列の累乗の計算が容易になる点がある。   
・正方行列ではない行列では、特殊な単位ベクトルがある場合に限り特異値分解を行うことができる。  
・特異値分解した画像のデータの行列から成分の小さい部分を取り除いていくと「画像はぼやけていく＝データ量を小さくできる」

# ラビットチャレンジ/科目「応用数学」/「第2章：確率・統計」  

## 100文字以上で要点のまとめ  

条件付き確率とは、  
・ある事象$X=x$が与えられた下で、$Y=y$となる確率
$$
P(Y=y|X=x) =\frac{P(Y=y,X=x)}{P(X=x)} 
$$

カンマは$\cap$と同じ意味

ベイズ則
$$
P(X=x|Y=y)P(Y=y) = P(Y=y,X=x) = P(X=x,Y=y) =  P(Y=y|X=x)P(X=x)
$$


事象$X$、確率変数$f(X)$、確率$P(X)$とする。  
・離散量（整数のような実数直線上の飛び飛びの値をイメージ）において
$$
期待値E(f) = \sum_{k=1}^{n} P(X = x_k)f(X = x_k)
$$
・連続量（実数全体の集合をイメージ）において
$$
期待値E(f) = \int P(X = x)f(X = x)dx
$$

分散は、  
・データの散らばり具合
・データの各々の値が、期待値からどれだけズレているのか平均したもの。

$$
\begin{align}
分散Var(f) &= E \Biggl( \biggl( f_{(X=x)} - E_{(f)} \biggl)^2 \Biggl)\\
&=E \biggl(f^2_{(X=x)} \biggl) - \biggl(E_{(f)} \biggl)^2
\end{align}
$$

共分散は、  
・2つのデータ系列の傾向の違い  
・正の値を取れば似た傾向、負の値を取れば逆の傾向、ゼロを取れば関係性に乏しい  

$$
\begin{align}
共分散Cov(f,g) &= E \Biggl( \biggl( f_{(X=x)} - E{(f)} \biggl) \biggl (g_{(Y=y)} - E{(g)} \biggl)\Biggl)\\
&=E (fg) - E{(f)}E{(g)}
\end{align}
$$

・分散の平方根を求めれば標準偏差  

相関係数は、    
・１に近ければ強い正の相関、0に近ければ相関なし、−１に近ければ強い負の相関  

$$
r = \frac{Cov(f,g)}{{\sqrt {Var(f)}}{\sqrt {Var(g)}}}
$$

ベルヌーイ分布  
・コイントスのイメージ  
・表と裏で出る割合が等しくなくとも扱える

$$
P(x| \mu) = \mu^x(1 - \mu)^{1-x}
$$

マルチヌーイ（カテゴリカル）分布  
・さいころを転がすイメージ  
・各面の出る割合が等しくなくとも扱える

二項分布  
・ベルヌーイ分布の多試行版

$$
P(x| \lambda,n) = \frac{n!}{x!(n - x)!} \lambda^x(1 - \lambda)^{n-x}
$$

正規分布（ガウス分布）  
・釣鐘型の連続分布

$$
N(x; \mu,\sigma^2) = \sqrt{{\frac{1}{2\pi \sigma^2}}}exp \Bigl(-\frac{1}{2\sigma^2}(x - \mu)^2 \Bigl)
$$

## 演習問題
袋の中に赤い玉3個と白い玉2個が入っている。赤い玉はA，B，Cの文字が，白い玉にはA，Bの文字が，それぞれ1個に対して1文字ずつ記されている。以下の問いに答えよ。

問5.2.1  出てきた玉が赤色であったとき，それに記されている文字がBである確率。

出てきた玉が赤玉である確率を$P(A)$、出てきたたまに記されている文字がBである確率を$P(B)$とすると、出てきた玉が赤玉であり、かつ、記された文字がBである確率は$P(A\cap B)$と表すことができる。

$$P(A) = \frac{3}{5}$$
$$P(A\cap B) = \frac{3}{5} \times \frac{1}{5} =  \frac{1}{5}$$

求める確率は
$$
P_A(B)= \frac{P(A\cap B)}{P(A)} =  \frac{1}{5} \div \frac{3}{5} = \frac{1}{3}
$$

※上記表記は高校数学「数学I」に基づいています。

## 100文字以上で要点のまとめ  

自己情報量

・対数の底が$2$のとき、単位はbit  
・対数の底がネイピアの$e$のとき、単位はnat

$$
I(x) = -log_e(P(x)) = log_e(W(x))
$$


シャノンエントロピー

・微分エントロピーとも呼ばれるが、微分しているわけではない。  
・自己情報量の期待値  

$$
\begin{align} H(x) &= E (I(x)) \\
& = -E(log(P(x))) \\
& = -\sum(P(x)log(P(x)))
\end{align}
$$

例）　１つのコイン投げを考える。  
表が出る事象をHead(H)、裏が出る事象をTail(T)とする。  
$P(H)= \frac{1}{2}$　表が出たことを知って得られる情報量   
$$
-logP(H)=-log_22^{-1} = 1(bit)
$$  
$P(T)= \frac{1}{2}$　裏が出たことを知って得られる情報量   
$$
-logP(T)=-log_22^{-1} = 1(bit)
$$

→ 実際にコインを投げる際にどの程度の情報量の増加が期待できるか？

$$
\begin{align} H(i) &=  -\sum_{i=T,H}(P(i)log(P(i)))
=-\frac{1}{2}log_2\frac{1}{2}--\frac{1}{2}log_2\frac{1}{2}= 1(bit)
\end{align}
$$

表、裏のどちらかが出るかが予めわかっている場合　$P(H)=1,P(T)=1$  
→ エントロピーは0

・2値エントロピー関数、複合事象のエントロピー、条件付きエントロピーなどにも後で目を通しておく。


カルバック・ライブラー ダイバージェンス(Kullback-Leibler divergence/KLダイバージェンス/KL情報量)

・２つの確率分布がどの程度似ているかを表す尺度  

・同じ事象・確率変数における異なる確率分布$P$, $Q$の違いを表す  

・同じ確率分布ではKLダイバージェンスは0になる

・KLダイバージェンスは常に0を含む正の値となり、確率分布が似ていない程、大きな値となる

$$
D_{KL}(P||Q) = \mathbb{E}_{X〜P} \biggl[log \frac{P(x)}{Q(x)} \biggl]
= \mathbb{E}_{x〜P}[log P(x) - log Q(x)]
$$

上記の式を書き換える

$$
I(Q(x)) - I(P(x)) = (-log(Q(x))) - (-log(P(x))) = log \frac{P(x)}{Q(x)}
$$

離散量Ver
$$
D_{KL}(P||Q) = \sum_{x}P(x)(-log(Q(x))) - (-log(P(x))) = \sum_{x}P(x)log \frac{P(x)}{Q(x)}
$$

連続量Ver
$$
{D_{KL}(P||Q) = \int_{-\infty}^{\infty}P(x)log \frac{P(x)}{Q(x)}dx}
$$

交差エントロピー

・KLダイバージェンスの一部分を取り出したもの

・Qについての自己情報量をPの分布で平均している


$$
D_{KL}(P||Q)=\sum_{x}P(x)(-log(Q(x)))-(-log(P(x)))
$$

$$
上記式のP(x)(-log(Q(x)) に注目して考える。
$$

$$
H(P,Q) = H(P) + D_{KL}(P||Q)
$$

$$
H(P,Q)=-\mathbb{E}_{x〜P}logQ(x) = -\sum_{x}P(x)logQ(x)
$$

KLダイバージェンスから必要なものだけを取り出した式により、効率的に解を求められる。

$$
